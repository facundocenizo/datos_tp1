```python
x = eventos['event'].value_counts().plot(kind='bar',figsize=(14,10),color = 'black')
x.set_xlabel('Eventos realizados',fontsize=12)
x.set_ylabel('Cantidad de eventos',fontsize=12)
x.set_title('Distribucion de cantidad de eventos por tipos', fontsize=15,color = 'blue')
plt.show()
```

```python
eventos['timestamp'].dt.year.value_counts()
```

```python
plot = eventos['timestamp'].dt.month.value_counts().plot(kind='bar',figsize=(14,10),color = 'black')
plot.set_xlabel('Mes',fontsize=12)
plot.set_ylabel('Cantidad de eventos',fontsize=12)
plot.set_title('Los meses con mas actividad (en los primeros 6 meses)', fontsize=15,color = 'blue')
plt.show()
```

```python
plot = eventos['timestamp'].dt.day.value_counts().head(5).plot(kind='bar',figsize=(14,10),color = 'black')
plot.set_xlabel('Los dias del mes con mas actividad',fontsize=12)
plot.set_ylabel('Cantidad de eventos',fontsize=12)
plot.set_title('Los dias del mes con mas actividad en el sitio', fontsize=15,color = 'blue')
plt.show()
```

```python
plot = eventos['timestamp'].dt.day.value_counts().tail(5).plot(kind='bar',figsize=(14,10),color = 'black')
plot.set_xlabel('Los dias del mes con menos actividad',fontsize=12)
plot.set_ylabel('Cantidad de eventos',fontsize=12)
plot.set_title('Los dias del mes con menos actividad en el sitio', fontsize=15,color = 'blue')
plt.show()
```

```python
plot = eventos['timestamp'].dt.day_name().value_counts().plot(kind='bar',figsize=(14,10),color = 'black')
plot.set_xlabel('Dias de la semana',fontsize=12)
plot.set_ylabel('Cantidad de eventos',fontsize=12)
plot.set_title('Distribucion de eventos durante la semana', fontsize=15,color = 'blue')
plt.show()
```

```python
eventosDeLaSemana = eventos.loc[((eventos['timestamp'].dt.day_name() == 'Monday') 
                                      | (eventos['timestamp'].dt.day_name() == 'Tuesday') 
                                      | (eventos['timestamp'].dt.day_name() == 'Wednesday')
                                      | (eventos['timestamp'].dt.day_name() == 'Thursday')
                                      | (eventos['timestamp'].dt.day_name() == 'Friday')
                                      
                                     ),:]['timestamp'].value_counts().sum()
```

```python
eventosDelFinDeSemana = eventos.loc[((eventos['timestamp'].dt.day_name() == 'Sunday') 
                                      | (eventos['timestamp'].dt.day_name() == 'Saturday') 
                                      
                                     ),:]['timestamp'].value_counts().sum()
```

```python
plt.figure(figsize=(8, 8))
sizes = [eventosDelFinDeSemana, eventosDeLaSemana]
nombres = ['eventoDelFinDeSemana', 'eventoSemanal']
plt.title('Cantidad de eventos dividos durante el fin de semana y los dias habiles de la semana', fontsize=20)
plt.pie(sizes, labels=nombres, autopct='%1.1f%%', startangle=20, colors=['red', 'lightblue'], explode=(0.1, 0))
plt.show()
```
