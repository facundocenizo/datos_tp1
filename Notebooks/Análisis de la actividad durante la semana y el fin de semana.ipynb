```python
import pandas as pd
import matplotlib.pyplot as plt
eventos = pd.read_csv("events.csv", low_memory=False)
eventos["timestamp"] = pd.to_datetime(eventos["timestamp"])
```

# Actividad en la semana y el fin de semana

```python
eventosDeLaSemana = eventos.loc[((eventos['timestamp'].dt.day_name() == 'Monday')
                                 | (eventos['timestamp'].dt.day_name() == 'Tuesday')
                                 | (eventos['timestamp'].dt.day_name() == 'Wednesday')
                                 | (eventos['timestamp'].dt.day_name() == 'Thursday')
                                 | (eventos['timestamp'].dt.day_name() == 'Friday')),:]['timestamp'].value_counts().sum()

eventosDelFinDeSemana = eventos.loc[((eventos['timestamp'].dt.day_name() == 'Sunday')
                                     | (eventos['timestamp'].dt.day_name() == 'Saturday')),:]['timestamp'].value_counts().sum()
```

```python
plt.figure(figsize=(8, 8))
sizes = [eventosDelFinDeSemana, eventosDeLaSemana]
nombres = ['Fin de semana', 'Semana']
plt.title('Cantidad de eventos', fontsize=20)
plt.pie(sizes, labels=nombres, textprops={'fontsize': 15}, autopct='%1.1f%%', startangle=20, colors=['orange','lightblue'], explode=(0.1, 0))
plt.show()
```

# Actividad por d√≠a

```python
eventos['timestamp'].dt.weekday.value_counts(sort=False)
```

```python
plot = eventos['timestamp'].dt.weekday.value_counts(sort=False).plot(kind='bar',figsize=(10,7),fontsize=12)
plot.set_xlabel('Dias',fontsize=14)
plot.set_ylabel('Cantidad de eventos',fontsize=14)
plot.set_title('Distribucion de eventos por dia', fontsize=20,color = 'blue')
plot.set_xticklabels(['Lunes', 'Martes', 'Miercoles', 'Jueves', 'Viernes', 'Sabado', 'Domingo'], rotation=0)
plt.show()
```
