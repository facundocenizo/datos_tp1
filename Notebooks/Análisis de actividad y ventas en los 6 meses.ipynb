```python
import pandas as pd
import matplotlib.pyplot as plt
eventos = pd.read_csv("events.csv", low_memory=False)
```

```python
compras = eventos.loc[eventos["event"]=="conversion"]
compras['timestamp'] = pd.to_datetime(compras['timestamp'])
compras["dia"] = compras.loc[:,"timestamp"].dt.date
compras = compras.loc[:,["timestamp", "dia"]]
compras = compras["dia"].value_counts()
```

```python
plot = compras.plot(kind='line',figsize=(12,8))
plot.set_title('Cantidad de ventas por día y por mes', fontsize=20,color = 'orange')
plot.set_xlabel('Meses',fontsize=20)
plot.set_ylabel('Cantidad de compras diarias',fontsize=20)
plot.set_xticklabels(['Enero','Febrero', 'Marzo','Abril', 'Mayo','Junio'])
plt.show()
```

```python
actividad = eventos.loc[:,["timestamp", "event"]]
actividad['timestamp'] = pd.to_datetime(actividad['timestamp'])
actividad["dia"] = actividad.loc[:,"timestamp"].dt.date
actividad = actividad["dia"].value_counts()
```

```python
plot = actividad.plot(kind='line',figsize=(12,8))
plot.set_title('Cantidad de eventos por día y por mes', fontsize=20,color = 'orange')
plot.set_xlabel('Meses',fontsize=20)
plot.set_ylabel('Cantidad de eventos diarios',fontsize=20)
plot.set_xticklabels(['Enero','Febrero', 'Marzo','Abril', 'Mayo','Junio'])
plt.show()
```
