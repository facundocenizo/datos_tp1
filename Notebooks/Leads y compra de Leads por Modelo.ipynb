```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
eventos = pd.read_csv('../Data/events.csv', low_memory=False)
```

```python
eventos.columns
```

```python
leadsPorPersona = eventos[eventos['event']=='lead']['person'].value_counts()
leadsPorPersona.head()
```

```python
leads = eventos[eventos['event'] == 'lead'][['person','model']]
cantLeadsPorModelo = leads.groupby('model').agg(np.size)

cantLeadsPorModelo = cantLeadsPorModelo[cantLeadsPorModelo['person'] > 3].sort_values(by='person')
display(cantLeadsPorModelo)
cantLeadsPorModelo.tail(10).plot(kind= 'barh')
plt.show()
```

```python
compras = eventos[eventos['event']=='conversion'][['model', 'person']]
```

```python
comprasSobreLeads = pd.merge(leads, compras, how='inner', on=['person', 'model'])
comprasSobreLeads.head()
```

```python
comprasSobreLeadsOrdenada = comprasSobreLeads.groupby('model').agg(np.size).sort_values(by='person', ascending=False)
comprasSobreLeadsOrdenada
```

```python
comprasSobreLeads['model'].value_counts().sort_values().tail(15).plot(kind='barh')
plt.show()
```

```python
juntas = pd.merge(cantLeadsPorModelo, comprasSobreLeadsOrdenada, how='inner', on=['model'])
juntas.columns = ['leads','conversion']
juntas
```

```python
juntas['no_conversion'] = juntas['leads'] - juntas['conversion']
juntas = juntas.sort_values(by='leads', ascending=False)
juntas
```

```python
# Make data: I have 3 groups and 7 subgroups
#group_names=['groupA', 'groupB', 'groupC']
group_names=juntas.index
#group_size=[12,11,30]
group_size=juntas['leads']

subgroup_names=['Si', 'No', 'Si', 'No', 'Si', 'No', 'Si', 'No', 'Si', 'No', 'Si', 'No', 'Si', 'No', 'Si', 'No']
subgroup_size=[6,25,14,12,2,19,13,7,2,13,1,8,1,4,2,2]
 
# Create colors
a, b, c, d ,e , f, g, h  = [plt.cm.Blues, plt.cm.Reds, plt.cm.Greens, plt.cm.Oranges,plt.cm.Purples,plt.cm.Greys,plt.cm.Wistia,plt.cm.RdPu]
 
# First Ring (outside)
fig, ax = plt.subplots()
ax.axis('equal')
mypie, _ = ax.pie(group_size, radius=2.3, labels=group_names, colors=[a(0.6), b(0.6), c(0.6), d(0.6),e(0.6), f(0.6), g(0.6), h(0.6)])
plt.setp( mypie, width=1.2, edgecolor='white')
 
# Second Ring (Inside)
mypie2, _ = ax.pie(subgroup_size, radius=1.8, labels=subgroup_names, labeldistance=0.80, colors=[a(0.5), a(0.4), b(0.5), b(0.4), c(0.5), c(0.4), d(0.5), d(0.4),e(0.5), e(0.4), f(0.5), f(0.4), g(0.5), g(0.4), h(0.5), h(0.4)])
plt.setp( mypie2, width=0.7, edgecolor='white')
plt.margins(0,0)

#plt.title('Cantidad de veces puesto en lead por modelo de telefono', fontsize=20)

# show it
plt.show()
```
