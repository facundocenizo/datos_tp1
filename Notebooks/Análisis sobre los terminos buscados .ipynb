```python
import numpy as np 
import pandas as pd

# plots
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns
%matplotlib inline
from wordcloud import WordCloud ## poner en consola 'python -m pip install wordcloud'


eventos = pd.read_csv('../Data/events.csv',low_memory=False)
```

```python
eventos['timestamp'] = pd.to_datetime(eventos['timestamp'])
```

```python
listaDeTerminosBuscados = eventos['search_term'].value_counts().reset_index()['index'].tolist()
```

```python
k = []
for i in listaDeTerminosBuscados:
    j = i.replace(' ','')
    j = j.lower()
    k.append(j)
    
listaDeTerminosBuscados = k
```

```python
text = ' '.join(listaDeTerminosBuscados)
```

```python
# Create the wordcloud object
wordcloud = WordCloud(width=1480, height=1480, background_color="green" ,margin=0,colormap="Reds", max_words=250).generate(text)
plt.figure(figsize=[14,19]) 
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

plt.show()

```
