```python
import numpy as np 
import pandas as pd

# plots
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns
%matplotlib inline
from wordcloud import WordCloud ## poner en consola 'python -m pip install wordcloud'


eventos = pd.read_csv('events.csv',low_memory=False)
```

```python
eventos['timestamp'] = pd.to_datetime(eventos['timestamp'])
```

```python
eventos["search_term"].value_counts()
```

```python
asd = eventos[["event", "search_term"]].dropna()
asd["search_term"].value_counts()
```

```python
wordcloud2 = WordCloud(width=1480, height=1480, background_color="green" ,margin=0,colormap="Reds", max_words=250).generate(' '.join(str(v) for v in asd["search_term"]))
wordcloud2
```

```python
plt.figure(figsize=[10,15]) 
# Display the generated image:
plt.imshow(wordcloud2, interpolation='bilinear')
plt.axis("off")

plt.show()
```
